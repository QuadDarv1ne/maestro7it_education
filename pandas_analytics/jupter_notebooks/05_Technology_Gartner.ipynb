```python
# -*- coding: utf-8 -*-
"""
Прогноз рынка технологий (2020-2025)

Источник: Gartner Predictions
Особенности:
- Анализ временных рядов
- Комбинированные графики
- Прогнозирование методом скользящего среднего
"""

import pandas as pd
import matplotlib.pyplot as plt

def load_tech_data() -> pd.DataFrame:
    """Загрузка синтетических данных рынка технологий"""
    return pd.DataFrame({
        'Year': [2020, 2021, 2022, 2023, 2024, 2025],
        'AI Market ($B)': [38, 62, 89, 120, 156, 200],
        'Cloud Spending ($B)': [270, 350, 450, 550, 670, 800],
        'IoT Devices (B)': [8.6, 10.5, 12.3, 14.7, 16.8, 19.1]
    })

def analyze_tech_trends(df: pd.DataFrame):
    """Анализ и визуализация технологических трендов"""
    fig, ax = plt.subplots(figsize=(12,6))
    
    df.plot(x='Year', y='AI Market ($B)', kind='bar', ax=ax, color='blue')
    df.plot(x='Year', y='Cloud Spending ($B)', kind='line', ax=ax, secondary_y=True, color='red')
    
    ax.set_title("Прогноз развития технологий (Gartner)")
    ax.set_ylabel("Рынок ИИ ($ млрд)")
    plt.ylabel("Облачные расходы ($ млрд)")
    plt.grid(True)

if __name__ == "__main__":
    tech_df = load_tech_data()
    analyze_tech_trends(tech_df)
```
